```python
gamma, epsilon = "", ""
zeros, ones = [0 for _ in range(len(inputs[0]))], [0 for _ in range(len(inputs[0]))]
for x in inputs:
    for i, n in enumerate(x):
        if n == "0":
            zeros[i] += 1
        else:
            ones[i] += 1
for z, o in zip(zeros, ones):
    if z > o:
        gamma += "0"
        epsilon += "1"
    else:
        gamma += "1"
        epsilon += "0"

print("puzzle 1:", int(gamma, 2) * int(epsilon, 2))
```

```python
def get_rating(inputs, i, rating="oxygen"):
    if len(inputs) == 1:
        return inputs[0]
    
    ones, zeros = [], []
    for x in inputs:
        if x[i] == "0":
            zeros.append(x)
        else:
            ones.append(x)
    
    i += 1
    if rating == "oxygen":
        if len(ones) >= len(zeros):
            return get_rating(ones, i, rating=rating)
        else:
            return get_rating(zeros, i, rating=rating)
    else: 
        if len(ones) >= len(zeros):
            return get_rating(zeros, i, rating=rating)
        else:
            return get_rating(ones, i, rating=rating)

oxygen, co2 = get_rating(inputs, 0, rating="oxygen"), get_rating(inputs, 0, rating="co2")
print("puzzle 2:", int(oxygen, 2) * int(co2, 2))
```
