```python
def x_split(spaces, ford):
    _spaces = []
    for i in range(ford):
        line = []
        for y, v in enumerate(spaces[i]):
            if v == "#" or spaces[-1-i][y] == "#":
                line.append("#")
            else:
                line.append(" ")
        _spaces.append(line)
    return _spaces

def y_split(spaces, ford):
    _spaces = []
    for line in spaces:
        _line = []
        for y in range(ford):
            if line[y] == "#" or line[-1-y] == "#":
                _line.append("#")
            else:
                _line.append(" ")
        _spaces.append(_line)
    return _spaces
```
