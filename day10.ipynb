```python
total = []
for line in inputs:
    cache = []
    for x in list(line):
        if x == '>':
            if cache.pop() != "<":
                break
        elif x == ']':
            if cache.pop() != "[":
                break
        elif x == ')':
            if cache.pop() != "(":
                break
        elif x == '}':
            if cache.pop() != "{":
                break
        else:
            cache.append(x)
    else:
        _total = 0
        for x in cache[::-1]:
            _total *= 5
            if x == '<':
                _total += 4
            elif x == '[':
                _total += 2
            elif x == '(':
                _total += 1
            elif x == '{':
                _total += 3
        total.append(_total)

print("puzzle 2:", sorted(total)[len(total) // 2])
```
