```python
def judge_6(signals, signal):
    return len((signals[8] - signals[1]) & signal) == 5

def judge_0(signals, signal):
    return len((signals[8] - signals[4] | signals[1]) & signal) == 5

def judge_9(signals, signal):
    return len((signals[4] | signals[7]) & signal) == 5

def judge_3(signals, signal):
    return len(signals[1] & signal) == 2

def judge_5(signals, signal):
    return len((signals[4] - signals[1]) & signal) == 2

def judge_2(signals, signal):
    return len((signals[8] - signals[4]) & signal) == 3
```

```python
total = 0
for line in inputs:
    signals_raw, outputs_raw =  line.split(' | ')
    signals = [0] * 10
    for signal in signals_raw.split():
        if len(signal) == 2:
            signals[1] = set(signal)
        elif len(signal) == 3:
            signals[7] = set(signal)
        elif len(signal) == 4:
            signals[4] = set(signal)
        elif len(signal) == 7:
            signals[8] = set(signal)

    for signal in signals_raw.split():
        signal = set(signal)
        if len(signal) == 6:
            if judge_6(signals, signal):
                signals[6] = signal
            elif judge_0(signals, signal):
                signals[0] = signal
            elif judge_9(signals, signal):
                signals[9] = signal
        if len(signal) == 5:
            if judge_3(signals, signal):
                signals[3] = signal
            elif judge_5(signals, signal):
                signals[5] = signal
            elif judge_2(signals, signal):
                signals[2] = signal

    signals = {tuple(sorted(list(s))): i for i, s in enumerate(signals)}
    # print(signals)
    
    _total = 0
    for output in outputs_raw.split():
        output = tuple(sorted(list(output)))
        # print(output)
        _total *= 10 
        _total += signals[output]
    
    # print(_total)
    # break
    total += _total

print("puzzle 2:", total)
```
