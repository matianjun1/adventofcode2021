```python
scanners = [[] for _ in range(30)]
n = -1
for x in inputs:
    if x.startswith("--"):
        n += 1
        continue
    if x == "":
        continue
    scanners[n].append([int(_x) for _x in x.split(",")])
```

```python
zero_map = set()
for x in scanners.pop(0):
    zero_map.add((x[0], x[1], x[2]))
zero_map
```

```python
def rotate(x, y, z):
    return [
        (x, y, z),
        (-y, x, z),
        (-x, -y, z),
        (y, -x, z),
        (x, -z, y),
        (x, -y, -z),
        (x, z, -y),
        (-z, y, x),
        (-x, y, -z),
        (z, y, -x),
        (-y, -z, x),
        (-y, -x, -z),
        (-y, z, -x),
        (-z, x, -y),
        (y, x, -z),
        (z, x, y),
        (-x, -z, -y),
        # (-x, y, -z),
        (-x, z, y),
        (-z, -y, -x),
        # (x, -y, -z),
        (z, -y, x),
        (y, -z, -x),
        # (y, x, -z),
        (y, z, x),
        (-z, -x, y),
        # (-y, -x, -z),
        (z, -x, -y),
    ]

points = [(0, 0, 0)]

def find(local):
    new_local = []
    for lx, ly, lz in local:
        new_local.append(rotate(lx, ly, lz))

    
    for i in range(24):
        _maps = {}
        for local in new_local:
            x, y, z = local[i]
            for zx, zy, zz in zero_map:
                point = (zx - x, zy - y, zz - z)
                _maps[point] = _maps.get(point, 0) + 1

            # print(_maps)
        for k, v in _maps.items():
            if v >= 12:
                points.append(k)
                # print(k)
                # print(local)
                # print(x, y, z)
                for local in new_local:
                    x, y, z = local[i]
                    point = (k[0] + x, k[1] + y, k[2] + z)
                    # print(point)
                    zero_map.add(point)
                return True
    
    return False
```
